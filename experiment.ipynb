 # Using "The Machine Learning" to find single transits
## An Experiment

In this notebook, I'm showing that we can find deep single transits by training a random forest classifier on injections into little chunks of (SAP) photometry.

To start, load the data files, split into months, and interpolate over missing data to get a consistent dataset.

```python
import kplr
import transit
import numpy as np
import matplotlib.pyplot as pl
```

```python
client = kplr.API()
```

```python
lcs = []
for lc in client.star(3558849).get_light_curves(short_cadence=False):
    # Load the data.
    data = lc.read()
    x = data["TIME"]
    y = data["PDCSAP_FLUX"]
    
    # Split into months.
    m = np.isfinite(x)
    gi = np.arange(len(x))[m]
    bi = np.arange(len(x))[~m]
    if len(bi):
        bi = bi[(bi > gi[0]) & (bi < gi[-1])]
        d = np.diff(bi)
        chunks = [slice(gi[0], bi[0])]
        for a, b in zip(bi[:-1][d > 1], bi[1:][d > 1]):
            chunks.append(slice(a+1, b-1))
        chunks.append(slice(bi[-1]+1, gi[-1]))
    else:
        chunks = [slice(gi[0], gi[-1])]
    
    # Interpolate missing data.
    for c in chunks:
        x0, y0 = x[c], y[c]
        m = np.isfinite(y0)
        if not np.any(m):
            continue
        y0[~m] = np.interp(x0[~m], x0[m], y0[m])
        lcs.append([x0, y0 / np.median(y0)])
```

```python
# We'll use windows with a half-width of 60 data points.
hw = 60

# Filter the light curve list to only include chunks that are long enough.
lcs = [l for l in lcs if len(l[0]) > 2*hw]
```

**Update**: what happens if we *remove* the month with this injection?

```python
lcs = lcs[:THE_TEST_LC] + lcs[THE_TEST_LC+1:]
```

Now we'll generate the training and validation set by choosing random 121 data point chunks to the light curves and injecting random physical transits into some of them.

```python
t = 0.5 / 24. * np.arange(-hw, hw+1)

# Generate some positive examples.
pos_sims = np.empty((10000, len(t)))
for j in range(len(pos_sims)):
    s = transit.System(transit.Central(q1=np.random.rand(), q2=np.random.rand()))
    s.add_body(transit.Body(period=np.exp(np.random.uniform(np.log(2e3), np.log(1e4))),
                            t0=np.random.uniform(-0.5, 0.5),
                            r=np.exp(np.random.uniform(np.log(0.03), np.log(0.3))),
                            b=np.random.rand()))    
    f = lcs[np.random.randint(len(lcs))][1]
    i = np.random.randint(hw, len(f)-hw) + np.arange(-hw, hw+1)
    pos_sims[j] = f[i] * s.light_curve(t)
    
# The negative examples are just random chunks of light curve without any injection.
neg_sims = np.empty((10000, len(t)))
for j in range(len(neg_sims)):
    f = lcs[np.random.randint(len(lcs))][1]
    i = np.random.randint(hw, len(f)-hw) + np.arange(-hw, hw+1)
    neg_sims[j] = f[i]
```

Now we'll format these samples for scikit learn, shuffle the order and split into training and validation sets.

```python
# Format.
X = np.concatenate((pos_sims, neg_sims), axis=0)
y = np.ones(len(X))
y[len(pos_sims):] = 0

# Shuffle.
inds = np.arange(len(X))
np.random.shuffle(inds)

# Split.
n = len(pos_sims)
X_train = X[inds[:n]]
y_train = y[inds[:n]]
X_valid = X[inds[n:]]
y_valid = y[inds[n:]]
```

Finally we'll train the (random forest) classifier.

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.grid_search import GridSearchCV 
```

```python
def pr_scorer(model, X, y):
    prec, rec, _ = precision_recall_curve(y, model.predict_proba(X)[:, 1])
    return auc(rec, prec)
```

```python
parameters = dict(
    n_estimators=[500, 1500, 2000],
    min_samples_leaf=[3, 5],
)

clf = ExtraTreesClassifier()
gs = GridSearchCV(clf, parameters, scoring=pr_scorer)
```

```python
gs.fit(X, y)
```

```python
gs.grid_scores_  
```

And evaluate it by testing the area under the ROC curve on the validation set.

```python
def roc_scorer(model, X, y):
    fpr, tpr, _ = roc_curve(y, model.predict_proba(X)[:, 1])
    return auc(fpr, tpr)


print("Train ROC = {0}".format(roc_scorer(clf, X_train, y_train)))
print("Valid ROC = {0}".format(roc_scorer(clf, X_valid, y_valid)))
print
print("Train PR = {0}".format(pr_scorer(clf, X_train, y_train)))
print("Valid PR = {0}".format(pr_scorer(clf, X_valid, y_valid)))
```

It's possible that we're over-fitting a bit. It's also possible that **this is a really easy problem**!!

Now, can we find that single transit that we injected at the beginning?

```python
# Choose a chunk of data.
t, f = TEST_LC_DATA  # map(np.array, lcs[THE_TEST_LC])
pl.plot(t, f, ".k")
pl.gca().axvline(THE_TEST_T0);
```

Now let's evaluate the model as a function of transit time.

```python
p = np.empty(len(t) - 2*hw)
for j, i in enumerate(range(hw, len(f)-hw)):
    p[j] = clf.predict_proba([f[i-hw:i+hw+1]])[0, 1]
```

And plot the resulting probability. The green line is the truth.

```python
pl.plot(t[hw:-hw], p, "k")
pl.gca().axvline(THE_TEST_T0, color="g", lw=2, alpha=0.5)
pl.xlabel("transit time")
pl.ylabel("probability");
```
